```python
def print_in_map(studios_1,times,colors,size):
    for i, row in studios_1[studios_1['gym_times'] == times].iterrows():
        label = row['gym_name']+ ": times -"+ str(row['gym_times'])
        folium.Marker(
            location=[row.latitude, row.longitude],
            tooltip=label,
            icon=folium.Icon(
                color=colors,
                icon_color="black",
                weight=1,
                fill_opacity="black"
                )
            ).add_to(m)
```
